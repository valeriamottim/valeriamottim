```python
import pandas as pd
import numpy as np

# Dados fornecidos
data = {
    "Data": ["15/01/2025", "15/01/2025", "15/01/2025", "15/01/2025", "15/01/2025",
             "15/01/2025", "15/01/2025", "15/01/2025", "15/01/2025", "15/01/2025"],
    "Cidade": ["São Paulo", "Rio de Janeiro", "Curitiba", "Porto Alegre", "Salvador",
               "New York", "London", "Tokyo", "Sydney", "Cape Town"],
    "Temperatura Máxima (°C)": [30.5, 35.0, 24.0, 28.0, 31.0,
                                5.0, 10.0, 12.0, 25.0, 22.0],
    "Temperatura Mínima (°C)": [22.0, 25.0, 18.0, 20.0, 24.5,
                                -3.0, 2.0, 6.0, 18.0, 16.0],
    "Precipitação (mm)": [12.0, np.nan, 8.0, 15.0, np.nan,
                          20.0, 5.0, 10.0, np.nan, 2.0],
    "Umidade Relativa (%)": [78, 70, np.nan, 82, 80,
                             60, 75, 65, 55, 68]
}

# Criação do DataFrame
df = pd.DataFrame(data)

# Exibir o DataFrame
print(df)
```

```python
# Substituir valores ausentes na coluna Precipitação pela média da Precipitação
df['Precipitação (mm)'].fillna(df['Precipitação (mm)'].mean(), inplace=True)

# Exibir o DataFrame
print(df)
```

```python
df = pd.DataFrame(data)
# Substituir valores ausentes na coluna Umidade Relativa pela mediana da Umidade Relativa
df['Umidade Relativa (%)'].fillna(df['Umidade Relativa (%)'].median(), inplace=True)

# Exibir o DataFrame
print(df)
```

```python
df = pd.DataFrame(data)

# Adicionar uma nova coluna chamada Amplitude Térmica
df['Amplitude Térmica (°C)'] = df['Temperatura Máxima (°C)'] - df['Temperatura Mínima (°C)']

# Exibir o DataFrame
print(df)
```
